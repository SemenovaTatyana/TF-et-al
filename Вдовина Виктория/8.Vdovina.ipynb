```python
import random
import math
from enum import Enum

pointNumberMin = 10 #минимальное количество точек
pointNumberMax = 15 #максимальное количество точек
pointMinXCoord = -100 #левая граница координаты x
pointMaxXCoord = 100 #правая граница координаты x
pointMinYCoord = -100 #левая граница координаты y
pointMaxYCoord = 100 #правая граница координаты y
linesByXRaise = True #строить ломаные слева направо, по возрастанию координаты x, а не в порядке расположения в списке
sortByAlphabet = False #сортировать цвета по алфавиту, а не по значениям
pointNum = random.randint(pointNumberMin, pointNumberMax) #количество точек
points = []

#класс перечисление цветов
class Color(Enum):
    RED = 1
    ORANGE = 2
    YELLOW = 3
    GREEN = 4
    LIGHT_BLUE = 5
    BLUE = 6
    VIOLET = 7

#класс для точек, содержит поля координат x и y, цвет точки
class Point:
    
    def __init__(self, x, y, color):

        self.x = x
        self.y = y
        self.color = color
    
    #вид объекта точка, при прямом выводе объекта   
    def __repr__(self):
        return "<%s, %s, %s>" % (self.x, self.y, self.color)
 
    
#подсчет длины линии между двумя точками
def countLength(pointA, pointB):
    xLength = math.fabs(pointA.x-pointB.x)
    yLength = math.fabs(pointA.y-pointB.y)
    return math.sqrt(xLength+yLength)

#ключ для сортировки по координате x
def byXKey(point):
    return point.x
#ключ для сортировки по цвету (в алфавитном или по значению цвета)
def byColorKey(point):
    if sortByAlphabet:
        return point.color.name
    else:
        return point.color.value

#подсчет длинны ломаной линии указанного цвета
def lineLength(color,points):
    
    length=0
    pointPrevious=None
    pointCurrent=None
    
    if linesByXRaise:
            points.sort(key=byXKey)
            
    for num in range(len(points)):
        if points[num].color.name==color:
            pointPrevious=pointCurrent
            pointCurrent=points[num]
            if pointPrevious is not None:
                length=length+countLength(pointPrevious, pointCurrent)
            
    return length
            
            
    

# генерирование случайных точек в указанном диапазоне и случайного цвета из заявленных
for num in range(pointNum):
    points.append(Point(random.randint(pointMinXCoord, pointMaxXCoord),random.randint(pointMinYCoord, pointMaxXCoord), random.choice(list(Color))))

#сортировка списка по цвету и вывод
points.sort(key=byColorKey)
for num in range(pointNum):
    print(points[num])

#вывод длин всех ломаных
for color in list(Color):
    print ("%s line length is: %s" % (color.name, lineLength(color.name,points)))
```
