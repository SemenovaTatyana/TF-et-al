```python
class Piece:
    
    def __init__(self,colour,name):
        self.name = name
        self.position = None
        self.colour = colour
    def valid(self,stpos,epos,colour,board):
        if epos in self.moves(stpos[0],stpos[1],board, colour = colour):
            return True
        return False
    def __repr__(self):
        return self.name
    
    def __str__(self):
        return self.name
    
    def moves(self,x,y,board):
        print('Ошибка: нельзя сходить')
        
    def answers(self,x,y,board, colour, ints):
        answers = []
        for xi,yi in ints:
            xt,yt = x+xi,y+yi
            while self.insidebounds(xt,yt):
                
                target = board.get((xt,yt),None)
                if target is None: 
                    answers.append((xt,yt))
                elif target.colour != colour: 
                    answers.append((xt,yt))
                    break
                else:
                    break
                
                xt,yt = xt + xi,yt + yi
        return answers
#проверка расположенности фигуры в пределах поля 
    def insidebounds(self,x,y):
        if x >= 0 and x < 8 and y >= 0 and y < 8:
            return True
        return False
#проверка отсутствия конфликтов    
    def conflict(self,board,initcolour,x,y):
        if self.insidebounds(x,y) and (((x,y) not in board) or board[(x,y)].colour != initcolour) : 
            return True
        return False
```

```python
cardinals = [(1,0),(0,1),(-1,0),(0,-1)]
diagonals = [(1,1),(-1,1),(1,-1),(-1,-1)]
```

```python
#проверка позиций для ладьи, чтобы не было конфликтов
def knight_list(x,y,i1,i2):
    return [(x+i1,y+i2),(x-i1,y+i2),(x+i1,y-i2),(x-i1,y-i2),(x+i2,y+i1),(x-i2,y+i1),(x+i2,y-i1),(x-i2,y-i1)]
```

```python
def king_list(x,y):
    return [(x+1,y),(x+1,y+1),(x+1,y-1),(x,y+1),(x,y-1),(x-1,y),(x-1,y+1),(x-1,y-1)]
```

```python
class Knight(Piece):
    def moves(self,x,y,board, colour = None):
        if colour is None : 
            colour = self.colour
        return [(xx,yy) for xx,yy in knight_list(x,y,2,1) if self.conflict(board, colour, xx, yy)]
```

```python
class Rook(Piece):
    def moves(self,x,y,board ,colour = None):
        if colour is None : 
            colour = self.colour
        return self.answers(x, y, board, colour, cardinals)
```

```python
class Bishop(Piece):
    def moves(self,x,y,board, colour = None):
        if colour is None : 
            colour = self.colour
        return self.answers(x, y, board, colour, diagonals)
```

```python
class Queen(Piece):
    def moves(self,x,y,board, colour = None):
        if colour is None : 
            colour = self.colour
        return self.answers(x, y, board, colour, cardinals+diagonals)
```

```python
class King(Piece):
    def moves(self,x,y,board, colour = None):
        if colour is None : 
            colour = self.colour
        return [(xx,yy) for xx,yy in king_list(x,y) if self.conflict(board, colour, xx, yy)]
```

```python
class Pawn(Piece):
    def __init__(self,colour,name,direction):
        self.name = name
        self.colour = colour
        self.direction = direction
    def moves(self,x,y,board, colour = None):
        if colour is None : 
            colour = self.colour
        answers = []
        if (x+1,y+self.direction) in board and self.conflict(board, colour, x+1, y+self.direction) : 
            answers.append((x+1,y+self.direction))
        if (x-1,y+self.direction) in board and self.conflict(board, colour, x-1, y+self.direction) : 
            answers.append((x-1,y+self.direction))
        if (x,y+self.direction) not in board and colour == self.colour : 
            answers.append((x,y+self.direction))
        return answers
```
